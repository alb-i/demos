```r
library(ggplot2)
library(tidyverse)
library(hrbrthemes)
library(viridis)
library(tseries)
```

```r
data <- read.csv("data/prep.csv")
```

```r
model <- lm(log(arr_flights) ~ as.factor(year) + as.factor(month) + extra_mon + extra_tue + extra_wed + extra_thu + extra_fri + extra_sat + extra_sun,data=data)
```

```r
summary(model)
```

```r
plot(model)
```

```r
y <- model$residuals

ggplot(data.frame(name=c("residuals"), value=c(y)), aes(x=name,y=value)) + geom_violin() +
    scale_fill_viridis(discrete = TRUE, alpha=0.6, option="A") +
    theme_ipsum() +
    theme(
      legend.position="none",
      plot.title = element_text(size=11)
    ) +
    ggtitle("Violin chart") +
    xlab("")
```
