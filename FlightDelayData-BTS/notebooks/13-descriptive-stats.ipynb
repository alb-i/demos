```python
control = data[data.t.isin(list(range(178,195)))]
```

```python
covid = data[data.t.isin(list(range(202,219)))]
```

```python
ctg = control.groupby([control.t,control.year,control.month,control.airport_name]).sum()
```

```python
cvg = covid.groupby([covid.t,covid.year,covid.month,covid.airport_name]).sum()
```

```python
N = max(len(ct_delay_ratio),len(cv_delay_ratio))
qs = [i/float(N) for i in range(0,N+1)]
N
```

```python
x = ct_delay_ratio
y = cv_delay_ratio

plt.figure()
plt.scatter(np.quantile(x,qs), np.quantile(y,qs),marker=',',lw=0,s=1)
plt.xlabel('control')
plt.ylabel('covid')
plt.show()
plt.close()
```

```python
x = ct_avg_delay[ct_avg_delay.isna()==False]
y = cv_avg_delay[cv_avg_delay.isna()==False]

N = max(len(x),len(y))
qs = [i/float(N) for i in range(0,N+1)]
print(f"N = {N}")

plt.figure()
plt.scatter(np.quantile(x,qs), np.quantile(y,qs),marker=',',lw=0,s=1)
plt.xlabel('control')
plt.ylabel('covid')
plt.show()
plt.close()
```

```python
x = ct_avg_delay[ct_avg_delay.isna()==False]
y = cv_avg_delay[cv_avg_delay.isna()==False]

N = max(len(x),len(y))
qs = [i/float(N) for i in range(0,N+1)]
print(f"N = {N}")

plt.figure(figsize=(8,6))
plt.scatter(qs, np.quantile(x,qs),color='blue',marker=',',lw=0,s=2)
plt.scatter(qs, np.quantile(y,qs),color='red',marker=',',lw=0,s=2)
plt.xlabel('quantile')
plt.ylabel('value')
plt.legend(["control","covid"])
plt.title("avg delay min")
plt.show()
plt.close()
```

```python
x = ct_delay_ratio
y = cv_delay_ratio

N = max(len(x),len(y))
qs = [i/float(N) for i in range(0,N+1)]
print(f"N = {N}")

plt.figure(figsize=(8,6))
plt.scatter(qs, np.quantile(x,qs),color='blue',marker=',',lw=0,s=2)
plt.scatter(qs, np.quantile(y,qs),color='red',marker=',',lw=0,s=2)
plt.xlabel('quantile (observations)')
plt.ylabel('value (fraction of delayed flights)')
plt.legend(["control","covid"])
plt.title("delayed flights fraction")
plt.show()
plt.close()
```
