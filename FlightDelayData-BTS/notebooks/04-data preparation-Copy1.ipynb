```python
prep['top30'] = prep.airport_name.isin(airport_top30).map({True:1, False:0})
```

```python
prep.isna().any(axis=1).sum()
```

```python
def calcDays(mask):
    def fn(row,mask=mask):
        yr,mt = row["year"],row["month"]
        if mt == 12:
            mt0,yr0 = 1, yr+1
        else:
            mt0,yr0 = mt+1, yr
        return np.busday_count("%4d-%02d"%(yr,mt),"%4d-%02d"%(yr0,mt0),weekmask=mask)
    return fn
prep['day_ct'] = prep.apply(calcDays('1111111'),axis=1)
for d in ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]:
    prep[d.lower()+"_ct"] = prep.apply(calcDays(d),axis=1)
    prep["extra_"+d.lower()] = prep[d.lower()+"_ct"] - 4
```

```python
prep["fulldata"] = prep.year.isin(list(range(2004,2022))).map({True:1, False:0})
```

```python
prep[prep.fulldata.eq(1)].year.unique()
```

```python
prep["t"] = prep.apply(lambda row: 12*(row["year"]-2003)+row["month"]-5, axis=1)
```

```python
prep["covid"] = prep.t.isin(list(range(202,219))).map({True:1,False:0})
```

```python
prep["control"] = prep.t.isin(list(range(178,195))).map({True:1,False:0})
```

```python
airport_indicator_list = "ATL,ORD,CLT,DEN,DFW,ADK,LAX,ITO,MSP,DTW,PHX,CDB,SEA,LAS,SLC,IAH,EKO,SFO,LIH,BWI,DCA,HIB,BTM,LGA,HNL,MCO,KOA,MDW,BOS,PHL,BJI,SAN,JFK,OGG,OTZ,BQN,LWS,SJC,GST,PIH,MIA,DAL,ELM,ABR,BNA".split(",")
```

```python
for a in airport_indicator_list:
    prep[a] = prep.airport.eq(a).map({True:1,False:0})
airportSums = prep.groupby([prep.t,prep.month,prep.year,prep.airport_name,prep.airport,prep.covid,prep.control,prep.extra_mon,prep.extra_tue,prep.extra_wed,prep.extra_thu,prep.extra_fri,prep.extra_sat,prep.extra_sun,prep.top30,prep.fulldata,prep.mon_ct,prep.tue_ct,prep.wed_ct,prep.thu_ct,prep.fri_ct,prep.sat_ct,prep.sun_ct]).sum()
for a in airport_indicator_list:
    airportSums[a] = (airportSums.__getitem__(a) > 0).map({True:1,False:0})

airportSums.to_csv('data/sum-airport-month.csv')
```
