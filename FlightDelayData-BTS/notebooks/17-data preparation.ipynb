```python
import pandas as pd
from datetime import date
import numpy as np
```

```python
prep = pd.read_csv("data/prep.csv",index_col=0)
prep
```

```python

monthSums = prep.groupby([prep.t,prep.month,prep.year,prep.covid,prep.control,prep.extra_mon,prep.extra_tue,prep.extra_wed,prep.extra_thu,prep.extra_fri,prep.extra_sat,prep.extra_sun,prep.top30,prep.fulldata,prep.mon_ct,prep.tue_ct,prep.wed_ct,prep.thu_ct,prep.fri_ct,prep.sat_ct,prep.sun_ct]).sum()

monthSums.to_csv('data/sum-months.csv')
```
